## 1. Explore the data
Loading the dataset and displaying basic information.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = 'StudentsPerformance.csv'
data = pd.read_csv(file)
data.head(5)
```

```python
df = pd.DataFrame(data)
pd.set_option('display.max_columns', None)
```

```python
# Get and print the column names
column_names = df.columns
print(column_names)
```

```python
# Create a new DataFrame with renamed columns
df_renamed = df.rename(columns={'STUDENT ID': 'ID', 'Student Age': 'Age', 
                                'Graduated high-school type': 'High-school type', 
                                'Regular artistic or sports activity': 'Activity',
                               'Do you have a partner': 'Existing partner',
                               'Total salary if available': 'Total salary',
                               'Transportation to the university': 'Transportation type',
                               'Accommodation type in Cyprus': 'Accommodation type',
                               'Mother’s education': 'Moms education',
                               'Father’s education ': 'Dads education',
                               'Number of sisters/brothers': 'Siblings amount',
                               'Mother’s occupation': 'Moms job', 'Father’s occupation': 'Dads job',
                               'Reading frequency.1': 'Scientific reading-f', 'Reading frequency': 'Reading-f',
                               'Attendance to the seminars/conferences related to the department': 'Seminar attendance',
                               'Impact of your projects/activities on your success': 'Projects impact on success',
                               'Attendance to classes': 'Class attendance', 'Preparation to midterm exams 1': 'Midterm 1 preparation',
                               'Preparation to midterm exams 2': 'Midterm 2 preparation', 'Taking notes in classes': 'Note taking',
                               'Discussion improves my interest and success in the course': 'Disscusion increases success',
                               'Cumulative grade point average in the last semester (/4.00)': 'Last semester grade point average',
                               'Expected Cumulative grade point average in the graduation (/4.00)': 'Expected grade point average'})
df_renamed.head(5)
```

```python
# Get and print the column names
column_names = df_renamed.columns
print(column_names)
```

```python
# Define a mapping from numbers to words
mapping = {1: 'married', 2: 'divorced', 3: 'died - one of them or both'}

# Replace the values in the 'Category' column using the mapping
df_renamed['Parental status'] = df_renamed['Parental status'].replace(mapping)
df_renamed['Projects impact on success'].replace({1: 'positive', 2: 'negative', 3:'neutral'}, inplace=True)
df_renamed['Class attendance'].replace({1: 'always', 2: 'sometimes', 3:'never'}, inplace=True)
df_renamed['Class attendance'].replace({'always':3, 'sometimes':2, 'never':1}, inplace=True)
df_renamed['Midterm 1 preparation'].replace({1: 'alone', 2: 'with friends', 3:'not applicable'}, inplace=True)
df_renamed['Midterm 2 preparation'].replace({1: 'alone', 2: 'with friends', 3:'not applicable'}, inplace=True)
df_renamed['Flip-classroom'].replace({1: 'not useful', 2: 'useful', 3:'not applicable'}, inplace=True)


df_renamed['High-school type'].replace({1: 'private', 2: 'state', 3: 'other'}, inplace=True)
df_renamed['Scholarship type'].replace({1: 'None', 2: '25%', 3: '50%', 4: '75%', 5: 'Full'}, inplace=True)
df_renamed['Transportation type'].replace({1: 'Bus', 2: 'Private car/taxi', 3: 'bicycle', 4: 'Other'}, inplace=True)
df_renamed['Accommodation type'].replace({1: 'rental', 2: 'dormitory', 3: 'with family', 4: 'Other'}, inplace=True)
df_renamed['Moms education'].replace({1: 'retired', 2: 'housewife', 3: 'government officer', 4: 'private sector employee', 5: 'self-employment', 6: 'other'}, inplace=True)
df_renamed['Dads education'].replace({1: 'retired', 2: 'government officer', 3: 'private sector employee', 4: 'self-employment', 5: 'other'}, inplace=True)

df_renamed['Activity'].replace({1: 1, 2: 0}, inplace=True)
df_renamed['Existing partner'].replace({1: 1, 2: 0}, inplace=True)
df_renamed['Seminar attendance'].replace({1: 1, 2: 0}, inplace=True)
df_renamed['Sex'].replace({1: 1, 2: 0}, inplace=True)
df_renamed['Additional work'].replace({1: 1, 2: 0}, inplace=True)

print(df_renamed['Activity'])
```
